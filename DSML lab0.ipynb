```python
import numpy as np

array = np.array([1,2,3,4,0])

indices = np.where(array > 2)
print("Indices where elements are greater than 2:", indices)

elements = array[indices]
print("Elements greater than 2:", elements)
```

```python
import numpy as np

sorted_array = np.array([10, 20, 30, 40, 50])

index = np.searchsorted(sorted_array, 23)
print("Index to insert 23:", index)

new_array = np.insert(sorted_array, index, 23)
print("New array after insertion:", new_array)
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["math", " Science", "English", "Computer", "Nepali", "Health"]

marks = [34,56,76,23,56,89]
plt.figure(figsize=(33,21))
plt.subplot(2, 2, 1)
plt.plot(subjects, marks, marker='o', color='green')
plt.title("Line Plot")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(16,8))
x = np.arange(len(subjects))
plt.scatter(x, marks, color='red')
plt.title("Scatter Plot")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.xticks(x, subjects)
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.bar(subjects, marks, color='purple')
plt.title("Bar Graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
plt.pie(marks, labels=subjects, startangle=90, colors=['green', 'blue', 'red', 'brown', 'gold', 'violet'])
plt.title("Pie Chart")
```
